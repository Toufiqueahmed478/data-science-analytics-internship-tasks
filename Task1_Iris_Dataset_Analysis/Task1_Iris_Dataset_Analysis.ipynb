# Task 1: Exploring and Visualizing the Iris Dataset

## Introduction and Problem Statement
The purpose of this task is to learn how to load a simple dataset, inspect its structure, summarize it, and create basic visualizations. In this notebook, we use the Iris dataset, which contains flower measurements for three iris species.


## Dataset Description
The Iris dataset contains measurements of iris flowers. The main columns are:
- sepal_length
- sepal_width
- petal_length
- petal_width
- species

We will use pandas for data handling, matplotlib and seaborn for visualization.


```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean visual style
sns.set(style="whitegrid")
```

## 1. Load the Dataset
We load the Iris dataset directly from seaborn.


```python
# Load Iris dataset
df = sns.load_dataset('iris')

# Display first five rows
df.head()
```

## 2. Dataset Structure
Now we check the shape, column names, and first few rows of the dataset.


```python
# Dataset shape: rows and columns
print("Dataset shape:", df.shape)

# Column names
print("Columns:", df.columns.tolist())

# First five rows
df.head()
```

## 3. Basic Summary
This helps us understand the numerical columns, average values, minimum, maximum, and spread of the data.


```python
# Summary statistics
df.describe()
```

```python
# Check missing values
df.isnull().sum()
```

## 4. Scatter Plot
Scatter plots help us see the relationship between two numerical variables. Here, we compare petal length and petal width.


```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species', s=80)
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```

### Scatter Plot Insight
The scatter plot shows that different iris species have different petal lengths and petal widths. Setosa is clearly separated from the other two species, while versicolor and virginica are closer to each other.


## 5. Histogram
Histograms show the distribution of values in a numerical column. Here, we check the distribution of sepal length.


```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='sepal_length', bins=20, kde=True)
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

### Histogram Insight
Most sepal length values are concentrated around the middle range. The histogram helps us understand how frequently different sepal length values appear in the dataset.


## 6. Box Plot
Box plots help us understand spread, median, and possible outliers. Here, we compare sepal width across different species.


```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='species', y='sepal_width')
plt.title('Sepal Width by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Width')
plt.show()
```

### Box Plot Insight
The box plot shows the spread of sepal width for each species. It also helps identify possible outliers and compare variation between species.


## 7. Conclusion
In this task, we loaded the Iris dataset, checked its structure using shape, columns, and head, summarized the data, and created three important visualizations: scatter plot, histogram, and box plot. The analysis shows that petal length and petal width are useful features for separating iris species, especially setosa.
